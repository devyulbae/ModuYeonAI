````python
import matplotlib.pyplot as plt
````

### 문제 1: 기본 라인 플롯 그리기

````python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title('Basic Line')
plt.show()
````

---

### 문제2 : 여러 개의 라인 플롯 그리기

````python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 4, 9, 16, 25]

plt.plot(x, y1, label='y=2x')
plt.plot(x, y2, label='y=x^2')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
````

---

### 문제 3: 산점도(Scatter plot) 그리기

````python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.scatter(x,y)
plt.show()
````

### 문제 4: 바 차트(Bar chart) 그리기

````python
categories = ['A', 'B', 'C', 'D', 'E']
values = [3, 7, 1, 5, 8]

plt.bar(categories, values)
plt.xlabel("categories")
plt.ylabel("value")
plt.show()
````

---

### 문제 5: 히스토그램(Histogram) 그리기

````python
data = [1, 1, 2, 3, 3, 3, 4, 5, 5, 5, 5]
print(set(data))

plt.hist(data)
plt.show()
````

---

### 문제 6: 파이 차트(Pie chart) 그리기

````python
sizes = [215, 130, 245, 210]
labels = ['A', 'B', 'C', 'D']

plt.pie(sizes, labels=labels, autopct="%1.1f%%")
# 원형을 유지하도록
plt.axis('equal')
plt.show()
````

---

### 문제 7: 박스 플롯(Box plot) 그리기

````python
data = [2, 3, 3, 4, 4, 4, 5, 6, 6, 7, 7, 7, 8, 8, 8, 9]

plt.boxplot(data)
plt.show()
````
